3. 3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
-> Dispersion refers to the spread or variability of data in a dataset — basically, how "spread out" the values are around the center (mean, median, etc.).

If data points are close to each other and the mean → low dispersion

If data points are far apart and vary a lot from the mean → high dispersion

Both variance and standard deviation are most useful when:

Data is quantitative

You want to understand consistency or volatility

You’re analyzing normal distributions

4.What is a box plot, and what can it tell you about the distribution of data?

-> A box plot is built around the five-number summary of a dataset:

1. Minimum – Smallest value (excluding outliers)

2. Q1 (First Quartile) – 25th percentile

3. Median (Q2) – 50th percentile (middle of the data)

4. Q3 (Third Quartile) – 75th percentile

5. Maximum – Largest value (excluding outliers)
    
     Example:
Let’s say you have exam scores out of 100: 45, 55, 60, 65, 70, 75, 80, 85, 90, 100

Min: 45

Q1: 60

Median: 72.5

Q3: 85

Max: 100

The box plot would show the middle 50% from 60 to 85, with the median near the middle, and maybe no outliers.

9.  Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule?
->  Key Properties of the Normal Distribution:
Symmetrical:


Mean = Median = Mode:


Total Area Under the Curve = 1:


Tails Extend Infinitely:

Shape is Defined by Two Parameters:

Mean (μ): Determines the center of the curve

Standard Deviation (σ): Determines the spread (how wide or narrow the curve is)

The Empirical Rule (68-95-99.7 Rule)
This rule describes how data is distributed in a normal distribution based on standard deviation:

Range from the Mean	Percent of Data	What it Means
μ ± 1σ (one SD)	~68%	About 68% of values fall within 1 SD of the mean

μ ± 2σ (two SDs)	~95%	About 95% of values fall within 2 SDs

μ ± 3σ (three SDs)	~99.7%	Almost all (99.7%) of values fall within 3 SDs.
